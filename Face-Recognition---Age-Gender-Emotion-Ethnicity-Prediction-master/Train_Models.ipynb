```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.layers import Dense
from keras.layers import MaxPooling2D
from keras.layers import Convolution2D
from keras.layers import Flatten
from keras.models import Sequential
```

## Train Model to check image is face or not

```python
data_frame = pd.read_csv('final_dataset.csv')
data_frame.head(10)
```

```python
data_frame['Emotion'].value_counts()
```

```python
def isFace(x):
    
    if x == 'Not_Face':
        return 0
    else:
        return 1
```

```python
X = data_frame.iloc[:,1:2501]
y = data_frame['Emotion'].apply(isFace)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model_isFace = Sequential()
model_isFace.add(Convolution2D(30,3, input_shape = (50,50,1), activation = 'relu', padding = 'same'))
model_isFace.add(MaxPooling2D(pool_size= (2,2)))
model_isFace.add(Convolution2D(40,3, activation = 'relu', padding = 'same'))
model_isFace.add(Convolution2D(45,3, activation = 'relu', padding = 'same'))
model_isFace.add(MaxPooling2D(pool_size= (2,2))) 
model_isFace.add(Flatten())
model_isFace.add(Dense(output_dim = 128, activation = 'relu'))
model_isFace.add(Dense(output_dim = 1, activation = 'sigmoid'))
```

```python
model_isFace.compile(optimizer='adam', loss = 'binary_crossentropy', metrics  = ['accuracy'])
```

```python
trainX = X_train.reshape(X_train.shape[0],50, 50, 1).astype( 'float32' )
testX = X_test.reshape(X_test.shape[0],50, 50, 1).astype( 'float32' )
```

```python
model_isFace.fit(trainX, y_train, batch_size=30, epochs = 10)
```

```python
score = model_isFace.evaluate(testX, y_test, batch_size=10)
```

```python
score
```

```python
model_json = model_isFace.to_json()
with open("face_model.json", "w") as json_file:
    json_file.write(model_json)
    
#serialise weights as HDF5

model_isFace.save_weights("face_weights.h5")
print("Saved")
```
